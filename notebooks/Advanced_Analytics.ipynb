 Interpretation

VaR measures the maximum expected loss at a given confidence level.

Lower VaR indicates lower downside risk.

 Task 2: Conditional Value at Risk (CVaR)

```python
var_df.head()
```

 Interpretation

CVaR measures the average loss during the worst 5% of market scenarios.

It provides a better estimate of tail risk than VaR.
